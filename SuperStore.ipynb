```python
#Select input and output from given dataset before apply sampling tech
X=df_new.drop("Category",axis=1)
Y=df_new["Category"]
```

```python
#train_test_split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=1,test_size=0.3)
```

```python
ros=RandomOverSampler()
```

```python
#apply scaling on X_train and X_test
#inuilt class StandardScaler use formula Z-score=(Xi-mean of (X))/std(X)
from sklearn.preprocessing import StandardScaler
#fit_transform on training data X_train and transform on Testing_data means X_test:
#Create object of StandardScaler class
ss=StandardScaler()

X_train=ss.fit_transform(X_train)
X_test=ss.transform(X_test)

print("After Scaling the training data")
print(X_train)
print("After Scaling the testing data")
print(X_test)
```

```python
#Before apply sampling technique: check Y_train and Y_test are imbalance or not
pd.Series(Y_train).value_counts()
```

```python
#Import Ramdom Over Sampling 
from imblearn.over_sampling import RandomOverSampler
```

```python
X_train_sample1,Y_train_sample1=ros.fit_resample(X_train,Y_train)
```

```python
X_train_sample1
```

```python
pd.Series(Y_train_sample1).value_counts()
```

```python
X_test_sample1,Y_test_sample1=ros.fit_resample(X_test,Y_test)
```

```python
pd.Series(Y_test_sample1).value_counts()
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
#This balance data is given to classification algorithim
#First:- LogisticRegression 
#First create a user difined function 
def create_model(model):
  model.fit(X_train_sample1,Y_train_sample1) #we train the model
  Y_pred=model.predict(X_test_sample1) #test
  print(classification_report(Y_test_sample1,Y_pred))
  print("Confusion Matrix :")
  #Confusion Matrix
  print(confusion_matrix(Y_test_sample1,Y_pred))
  return model
```

***Deep Learning***

```python
import tensorflow as tf
from keras import regularizers
```

```python
ss=StandardScaler()
X_train=ss.fit_transform(X_train)
X_test=ss.transform(X_test)
```

```python
df_new["Category"].dtypes
```

```python
df_new.columns
```

```python
df_new.shape
```

```python
#Create the neural network
#model=tf.keras.Sequential([tf.keras.layers.Dense(units=25,activation="relu",input_dim=2),
                       #tf.keras.layers.Dense(units=3,activation="softmax")])
```

```python
model=tf.keras.Sequential([tf.keras.layers.Dense(units=10,activation="relu",input_shape=(X.shape[1],)),
                       tf.keras.layers.Dense(units=10,activation="relu"),
                       tf.keras.layers.Dense(units=10,activation="relu"),
                       tf.keras.layers.Dense(units=3,activation="softmax")])
```

```python
print("Neural Network Model Summary")
print(model.summary())
```

```python
#Compile the model
model.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=['accuracy'])
#Here show losses and score both (score means show accuracy of model)
```

```python
#train the model
trained_model=model.fit(X_train,Y_train,epochs=200,batch_size=20)
```

```python
plt.plot(trained_model.history["loss"])
plt.show()
```

```python
plt.plot(trained_model.history["accuracy"])
plt.show()
```

```python
plt.plot(trained_model.history["loss"])
plt.plot(trained_model.history["accuracy"])
plt.show()
```

```python
#Show loss and accuracy on testing data, use inbuilt method evaluate()
model.evaluate(X_test,Y_test)
```

```python
#Show loss and accuracy on training data, use inbuilt method evaluate()
model.evaluate(X_train,Y_train)
```

```python
#Testing the model
Y_pred=model.predict(X_test)
```

```python
print(Y_pred)
```

```python
#Select class of maximumn prob. for each sample
Y_pred=Y_pred.argmax(axis=1)
print(Y_pred)
```

```python
print(classification_report(Y_pred,Y_test))
print(confusion_matrix(Y_pred,Y_test))
```
